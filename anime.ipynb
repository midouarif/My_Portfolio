```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
pd.set_option('display.max_columns', None)
```

```python
anime = pd.read_csv(r"C:\Users\surface\Downloads\anime.csv\anime.csv")
anime.head()
```

```python
anime.info()
```

```python
column_names = anime.columns.tolist()
print(column_names)
```

```python
drop_anime = ['real_start_date','real_end_date']
```

```python
anime = anime.drop(drop_anime,axis=1)
```

```python
anime['genres'].info()
```

```python
anime[anime['start_date'] != anime['real_start_date']]
```

```python
anime
```

```python
anime['episode_duration'] = anime['episode_duration'].astype('str')
anime['total_duration'] = anime['total_duration'].astype('str')
```

```python
anime['episode_duration'] = anime['episode_duration'].apply(lambda x: x.replace('0 days',''))
anime['total_duration'] = anime['total_duration'].apply(lambda x: x.replace('0 days',''))
```

```python
anime
```

```python
anime['start_date'] = anime['start_date'].astype('datetime64[ns]')
anime['end_date'] = anime['end_date'].astype('datetime64[ns]')

```

```python
anime.info()
```

```python
t = ['type','status','episodes','source','rating','broadcast_day']
```

```python
for i in t :
    bins = anime[i].nunique()
    #plt.figure(figsize=(25, 5))
    sns.histplot(data=anime, x=i, bins=bins)
    plt.title(i)
    plt.xticks(rotation=90)
    plt.show()
```

```python
anime.groupby('episodes')['anime_id'].count().sort_values(ascending=False).head(50).plot(kind = 'bar',figsize=(15,5))
```
